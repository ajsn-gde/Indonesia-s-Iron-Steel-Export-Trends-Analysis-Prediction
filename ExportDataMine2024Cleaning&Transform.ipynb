# Export

##Library

```python
import pandas as pd
import numpy as np
import re
```

## Access Data

```python
data = pd.read_excel('/content/drive/MyDrive/Data Science/Dataset/exim_https____exim_Kamis, 20 Februari 2025.xlsx', skiprows=1, header=[0, 1, 2, 3])
data
```

## Transform & Clean Data

```python
data = data.iloc[:-1, 1:-1]
data
```

```python
#mengganti jadi 0 untuk setiap header yang memiliki kata Unnamed
data.columns = [
    tuple("0" if "Unnamed" in str(x) else str(x) for x in col)
    for col in data.columns
]
```

```python
#melakukan join dari
data.columns = [
    '_'.join([str(x) for x in col if str(x) != "0"])  # Abaikan "0"
    for col in data.columns
]
```

```python
data
```

```python
data = data.drop(columns="ctr_pod_bulan")
```

```python
data
```

```python
data_melted = pd.melt(data, id_vars='kodehs', value_vars=data.iloc[:,1:], var_name='Pod_Month', value_name='Revenue')
```

```python
data_melted['Null'] = data_melted['Revenue'].isna()
data_melted = data_melted[data_melted["Null"] == False]
```

```python
data_melted = data_melted.drop(columns="Null")
data_melted.reset_index(drop=True, inplace=True)
data_melted
```

```python
data_melted[['Country', 'Pod', 'rMonth']] = data_melted['Pod_Month'].str.split('_', expand=True)
```

```python
data_melted
```

```python
data_melted.drop(columns='Pod_Month')
```

```python
data_melted[['HS Code', 'Name']] = data_melted['kodehs'].str.split(' ',n=1, expand=True)
```

```python
data_melted[['cMonth','Month']] = data_melted['rMonth'].str.split(' ', n=1, expand=True)
```

```python
data_melted = data_melted.drop(columns=['kodehs', 'Pod_Month', 'cMonth', 'rMonth'])
```

```python
data_melted
```

```python
data_melted = data_melted.reindex(columns=['HS Code', 'Name', 'Country', 'Pod', 'Month', 'Revenue'])
data_melted
```

```python
match = []
for i in data_melted['HS Code']:
  match.append(re.search(r'\d+', i))
data_melted['HS Code'] = [x.group() for x in match]
```

```python
data_melted['HS Code'].unique()
```

```python
data_melted
```

## Result

```python
data_melted.to_excel('export2024hscode71_83.xlsx')
```

# Import

```python
data_import = pd.read_excel('/content/drive/MyDrive/Data Science/Dataset/exim_https____exim_Jumat, 21 Februari 2025.xlsx', skiprows = 1, header = [0, 1, 2, 3])
```

```python
data_import = data_import.iloc[:-1, 1:-1]
data_import
```

```python
data_import
```

```python
data_import.columns = [
    '_'.join([str(x) for x in col if str(x) != "0"])
    for col in data_import.columns
]
```

```python
data_import = data_import.drop(columns="ctr_pod_bulan")
```

```python
data_import_melted = pd.melt(data_import, id_vars='kodehs', value_vars=data_import.iloc[:,1:], var_name='Ctr_Pod_Month', value_name='Revenue')
```

```python
data_import_melted["Null"] = data_import_melted['Revenue'].isna()
data_ic = data_import_melted[data_import_melted["Null"]!=True]
data_ic = data_ic.drop(columns="Null")
data_ic.reset_index(drop=True, inplace=True)
data_ic
```

```python
data_ic[['Country', 'Pod', 'rMonth']] = data_ic['Ctr_Pod_Month'].str.split('_', expand=True)
data_ic[['cMonth','Month']] = data_ic['rMonth'].str.split(' ', n=1, expand=True)
data_ic[['kode hs', 'Product']] = data_ic['kodehs'].str.split(' ', n=1, expand=True)
data_ic = data_ic.drop(columns=['Ctr_Pod_Month', 'cMonth', 'rMonth','kodehs'])
data_ic
```

```python
data_ic = data_ic.reindex(columns=['kode hs', 'Product','Country', 'Pod', 'Month', 'Revenue'])
data_ic
```

```python
match = []
for i in data_ic['kode hs']:
  match.append(re.search(r'\d+', i))
data_ic['HS Code'] = [x.group() for x in match]
```

```python
data_ic = data_ic.drop(columns='kode hs')
data_ic = data_ic.reindex(columns=['HS Code','Product','Country', 'Pod', 'Month', 'Revenue'])
data_ic
```

```python
data_ic.to_excel('import2024hscode71_83.xlsx')
```
